```python
rope = [(0, 0)] * 10
tail = [rope[-1]]

for i in inp:
    d, s = i.split()
    s = int(s)
    for _ in range(s):
        xh, yh = rope[0][0], rope[0][1]
        if d == "U":
            rope[0] = (xh, yh+1)
            
        elif d == "D":
            rope[0] = (xh, yh-1)
            
        elif d == "R":
            rope[0] = (xh+1, yh)
            
        elif d == "L":
            rope[0] = (xh-1, yh)
        
        for j in range(1, len(rope)):
            r = rope[j - 1][0] - rope[j][0]
            c = rope[j - 1][1] - rope[j][1]

            if r != 0:
                r -= 1 if r > 0 else -1

            if c != 0:
                c -= 1 if c > 0 else -1

            if r or c:
                rope[j] = (rope[j - 1][0] - r, rope[j - 1][1] - c)

        tail.append(rope[-1])

            
temp = []
for x in tail:
    if x not in temp:
        temp.append(x)

len(temp)
```
