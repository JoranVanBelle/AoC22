# Functions

```python
def convert_to_int(value):
    try:
        return int(value)
    except ValueError:
        return value
```

# Input

```python
inp = []
correct = 0
with open("data/dag05.txt", "r") as f:
    text = f.read()

inp = text.split("\n")
for i in range(len(inp)):
    inp[i] = inp[i].replace("    ", "[.]")
    # print(inp[i])
```

# Part 1

```python
temp = []
amount_of_rows = 0
string = ""
counter = -1
command = []

for i in inp:
    # Amount of columns of crates
    amount_of_rows += 1
    if i.startswith(" 1 "):
        amount_of_cols = int(i[-2])
        break

matrix = [[] for i in range(amount_of_cols)]

for i in inp:
    # Crates
    if "[" in i:
        temp_string = ""
        temp_ls = []
        write = False
        
        for j in i:
            if j == "[":
                write = True
            if write and j != "[" and j != "]":
                temp_string += j
            if j == "]":
                write = False
        counter += 1
        matrix[counter].append(list(temp_string))


# Get the commands
for i in inp:


    if "move" in i:
        temp = i.split(" ")
        for j in range(len(temp)):
            temp[j] = convert_to_int(temp[j])
        
        for j in temp:
            if type(j) == int:
                command.append(j)

flattend = []
for i in matrix:
    if i != []:
        flattend.append(i[0])

transposed = [[flattend[j][i] for j in range(len(flattend))] for i in range(len(flattend[0]))]

for i in transposed:
    while "." in i:
        i.remove(".")

for i in range(0, len(command)-2, 3):
    amount_of_crates = command[i]
    start_crate = command[i+1] - 1
    end_crate = command[i+2] - 1

    # print(f"amount_of_crates: {amount_of_crates}, start_crate: {start_crate}, endcrate: {end_crate}")

    # Write logica to switch the crates from place
    # print(f"Current state: {transposed}")

    crates_tebe_moved = transposed[start_crate][:amount_of_crates]
    # print(f"crates_tebe_moved: {crates_tebe_moved}")
    if len(crates_tebe_moved) > 1:
        crates_tebe_moved[::-1]

    for elem in crates_tebe_moved:
        transposed[end_crate].insert(0, elem)
    transposed[start_crate] = transposed[start_crate][amount_of_crates:]

result = ""
for i in transposed:
    result += i[0]
    
result
```
