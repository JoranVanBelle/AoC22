# Part 1

```python
inp = []
with open("data/dag1.txt", "r") as f:
    for r in f:
        inp.append(r.rstrip())

temp_list = []
temp = 0
for i in inp:
    if i != "":
        temp += int(i)
    else:
        temp_list.append(int(temp))
        temp = 0

print((max(temp_list), temp_list.index(max(temp_list))))
```

# Part 2

```python
def maxi(temp_list):
    return max(temp_list)

AMOUNT_OF_ELVES = 3

result = 0
for i in range(AMOUNT_OF_ELVES):
    val = maxi(temp_list)
    result += val
    temp_list.remove(val)

result
```
